# Compute rectangle area

```python
def compute_rectangle_area(height, width):
    '''
    This function aims to compute area for a rectangle.
    
    height -- the height of the rectangle
    width -- the width of the rectangle   
    
    This function returns the area of the rectangle
    '''
    
    area = height*width
    
    return area
```

```python
def compute_rectangle_area(height=0, width=0):
    '''
    This function aims to compute area for a rectangle.
    
    height -- the height of the rectangle
    width -- the width of the rectangle   
    
    This function returns the area of the rectangle
    '''
    
    area = height*width
    
    return area
```

# Sigmoid function 

```python
import math

def compute_sigmoid(x):
    '''
    this function aims to compute the sigmoid value of a number
    
    x -- input number
    
    Output is sigmoid value
    '''
    
    result = 1 / (1 + math.exp(-x))
    return result
    
# sample data: x=0 --> y=0.5
# sample data: x=5 --> y=0.99


x = 0.5
result = compute_sigmoid(x)
print(result)
```

```python
# Plot 
import matplotlib.pyplot as plt
import numpy as np

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Create a range of values for x from -10 to 10 with 100 points
x = np.linspace(-10, 10, 100)

# Calculate the sigmoid values for each x
y = sigmoid(x)

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='b')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0.5, color='r', linestyle='--', linewidth=0.8)  # Plot a horizontal line at y=0.5
plt.axvline(x=0, color='g', linestyle='--', linewidth=0.8)    # Plot a vertical line at x=0
plt.show()
```

# newton_method

```python
def newton_method(N, num_loops):
    '''
    This function aims to compute square root for the number N
    
    N -- the number needs to take the square root
    num_loops -- number of loops used for this optimization
    '''
    
    x_n = N/2
    
    for i in range(num_loops):
        x_np1 = (x_n + N/x_n) / 2
        x_n = x_np1
        
    return x_np1

print(newton_method(4, 100))
print(newton_method(2, 100))
```
